# Clustering

Clustering is used in a wide variety of applications, including:

* customer segmentation
* data analysis
* dimensionality reduction  
*  anomaly detection
* semi-supervised learning
* information retrival
* image segmentation

Despite all these applications, there is not a unifying definition of cluster,
it really depends on the context.

Different algorithms produce different clustering solutions.


For instance, there are algorithms based on the notion of centroid, 
whereas other algorithms focus on regions densely populated.

There are also algorithms that look for a hierarchy of clusters.

Il Clustering è una tecnica con approccio di apprendimento non supervisionato, a differenza della Classification che sfrutta un apprendimento supervisionato.

Nel clustering il modello riceve dei dati senza etichette e cerca di dividerli in gruppi, un esempio di utilizzo è la customer segmentation:

Clienti che comprano spesso

Clienti che comprano poco

Clienti interessati a prodotti economici

Clienti interessati a prodotti costosi

Può essere usato anche per ridurre la dimensionalità,cioè invece di rappresentare ogni dato con tantissime feature, posso raggruppare dati simili e semplificare il problema.

Non esiste una definizione unica di cluster.

Un cluster è, in generale, un gruppo di elementi simili, però il significato preciso dipende dal problema

## K-Means

Let's start by generating some blobs:

Un blob è un gruppo di punti generato artificialmente.

L'idea del K-Means è: ho dei punti,voglio dividerli in k gruppi,ogni gruppo ha un centro, chiamato centroide,ogni punto viene assegnato al centroide più vicino.

```python
from sklearn.datasets import make_blobs
```

```python
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])
#blob_centers sono i centri veri dei gruppi,
# ogni riga è un centro di un cluster
#Quindi stiamo dicendo che voglio generare punti attorno a questi 5 centri.
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])
#in blob_std definiamo la deviazione standard di ogni blob
```

```python
X, y = make_blobs(n_samples=2000, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)
#qui stiamo generando davvero i dati.
#X = coordinate dei punti generati
#y = etichetta del blob da cui ogni punto è stato generato
#n_samples=2000, indica che dovranno essere generati 2000 esempi, 
#quindi X avrà 2000 righe
'''
Attenzione: nel clustering normalmente y non la useresti per addestrare.
Qui esiste solo perché stiamo generando dati artificiali e 
sappiamo da quale blob arriva ogni punto.
'''
```

Let's see the ``KMeans`` algorithm.

It tries to find each blob center, then each instance will be assigned to a blob center.

```python
from sklearn.cluster import KMeans
```

Note that you have to specify the number of clusters $k$ the algorithm has to find. 

In general, choosing the right value for $k$ is not  easy or obvious.

K-Means prova a trovare il centro di ogni blob.

Poi assegna ogni istanza, cioè ogni punto, al centroide più vicino.

Essenzialmente fa :

1. scelgo k centri

2. assegno ogni punto al centro più vicino

3. ricalcolo il centro di ogni gruppo

4. ripeto finché i centri non cambiano più molto.

```python
'''
Create KMeans object with random_state = 42 and n_clusters = 5
'''

k = 5 # k è il numero di cluster(gruppi), ricordiamo che anche il numero
#dei Blob è 5.
#Un blob non è un centroide, è un insieme di punti
#questo è uno dei problemi del k-Means, ovvero definire a priori il 
#numero di cluster
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X) #X contiene 2000 punti
#nella .fit non si usa y,pk K-Means è unsupervised.
y_pred = kmeans.predict(X)
#y_pred contiene numeri da 0 a 4, perché abbiamo 5 cluster
```

### Decision Boundaries
If you plot the cluster’s decision boundaries, you get a Voronoi tessellation.

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='blue'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=10, linewidths=10,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
               marker='x', s=9, linewidths=10,
               color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(8, 4))
plot_decision_boundaries(kmeans, X)
save_fig("voronoi_plot")
plt.show()
```

Not bad! Some of the instances near the edges were probably assigned to the wrong cluster, but overall it looks pretty good.

### Hard Clustering _vs_ Soft Clustering

Rather than arbitrarily choosing the closest cluster for each instance, 
which is called _hard clustering_, 
it might be better measure the distance of each instance to all 5 centroids. 

This is what the `transform()` method does:

```python
kmeans.transform(X_new)
#.transform calcola le distanze dai centroidi
#kmeans.transform(X_new) calcola la distanza di ogni nuovo punto 
#da ogni centroide, X_new contiene 4 punti e i cluster sono 5, 
# l’output sarà una matrice:(4, 5), ovvero 4 righe = 4 nuovi punti e 
#5 colonne = distanza da ciascuno dei 5 centroidikmeans.transform(X_new)
'''
Esempio
[
  [distanza punto 1 da cluster 0, distanza punto 1 da cluster 1, ..., distanza punto 1 da cluster 4],
  [distanza punto 2 da cluster 0, distanza punto 2 da cluster 1, ..., distanza punto 2 da cluster 4],
  [distanza punto 3 da cluster 0, distanza punto 3 da cluster 1, ..., distanza punto 3 da cluster 4],
  [distanza punto 4 da cluster 0, distanza punto 4 da cluster 1, ..., distanza punto 4 da cluster 4]
]
La colonna con la distanza più piccola indica il cluster scelto da .predict
'''

#kmeans.predict(X_new) sceglie l’indice della distanza minima 
# di ogni riga di kmeans.transform(X_new)
```

1. K-Means può assegnare nuovi punti a un cluster con predict().

2. I confini decisionali di K-Means dividono lo spazio in regioni di Voronoi.

3. transform() non restituisce il cluster, ma le distanze da tutti i centroidi.

La differenza più importante è:

predict(X_new)
= hard clustering
= dammi un solo cluster per ogni punto

transform(X_new)
= soft information
= dammi le distanze da tutti i centroidi

kmeans.predict(X_new) 
risponde alla domanda: a quale cluster appartiene ogni punto?

kmeans.transform(X_new) risponde alla domanda:

You can verify that this is indeed the Euclidian distance between each instance and each centroid:

### K-Means Algorithm

The K-Means algorithm is one of the fastest and simplest clustering algorithms.
* First initialize $k$ centroids randomly: $k$ distinct instances are chosen randomly from the dataset and the centroids are placed at their locations.
* Repeat until convergence (i.e., until the centroids stop moving):
    * Assign each instance to the closest centroid.
    * Update the centroids to be the mean of the instances that are assigned to them.

Il K-Means prima inizializza casualmente k centroidi,all’inizio K-Means non sa dove siano i gruppi, quindi sceglie dei centroidi iniziali in modo casuale. Quindi se k = 5, prende 5 punti casuali tra i 2000 punti di X.

**Primo Passo:**

 assegna ogni istanza di X al centroide più vicino.

Una istanza è una riga di X, cioè un punto.

Quindi per ogni punto K-Means fa:

calcolo la distanza dal centroide 0

calcolo la distanza dal centroide 1

calcolo la distanza dal centroide 2

calcolo la distanza dal centroide 3

calcolo la distanza dal centroide 4

assegno il punto al centroide più vicino

Questo è il passo di assegnazione.

**Secondo passo:**

ricalcola i centroidi come media dei punti assegnati a ciascun cluster

Esempio: Cluster 0 contiene questi punti:

[1, 2]

[2, 2]

[3, 5]

Il nuovo centroide del cluster 0 sarà la media:

x1 medio = (1 + 2 + 3) / 3 = 2

x2 medio = (2 + 2 + 5) / 3 = 3

nuovo centroide = [2, 3]

Quindi il centroide si sposta verso il centro dei punti che gli sono stati assegnati.

SCHEMA:

1. scelgo k centroidi iniziali a caso

2. assegno ogni punto al centroide più vicino

3. ricalcolo ogni centroide come media dei punti assegnati

4. ripeto i passi 2 e 3

5. mi fermo quando i centroidi non cambiano più




The `KMeans` class applies an optimized algorithm by default. 
To get the original K-Means algorithm (for educational purposes only), you must set `init="random"`, `n_init=1`and `algorithm="full"`. 
These hyperparameters will be explained below.

### K-Means Variability
Unfortunately, although the algorithm will eventually converge to a solution, there is no guarantee it is the optimal solution.
It strongly depends on the centroid initialization.


In the original K-Means algorithm, the centroids are just initialized randomly, and the algorithm simply runs a single iteration to gradually improve the centroids, as we saw above.

However, one major problem with this approach is that if you run K-Means multiple times (or with different random seeds), it can converge to very different solutions, as you can see below:

K-Means può dare risultati diversi in base ai centroidi iniziali.

Purtroppo, anche se l’algoritmo prima o poi converge a una soluzione, non è garantito che sia la soluzione ottima.

Cioè K-Means si ferma quando i centroidi non si muovono più, però questo non significa per forza che abbia trovato il clustering migliore possibile.

Può trovare una soluzione **locale** e non **globale** 

```python
def plot_clusterer_comparison(clusterer1, clusterer2, X, title1=None, title2=None):
    clusterer1.fit(X)
    clusterer2.fit(X)

    plt.figure(figsize=(10, 3.2))

    plt.subplot(121)#121 significa 1 riga,2 colonne,grafico numero 1
    plot_decision_boundaries(clusterer1, X)
    if title1:
        plt.title(title1, fontsize=14)

    plt.subplot(122)
    plot_decision_boundaries(clusterer2, X, show_ylabels=False)
    if title2:
        plt.title(title2, fontsize=14)
```

```python
kmeans_rnd_init1 = KMeans(n_clusters=5, init="random", n_init=1,
                         random_state=11, max_iter=2)
kmeans_rnd_init2 = KMeans(n_clusters=5, init="random", n_init=1,
                         random_state=200, max_iter=2)

plot_clusterer_comparison(kmeans_rnd_init1, kmeans_rnd_init2, X,
                          "Solution 1", "Solution 2 (with a different random init)")

save_fig("kmeans_variability_plot")
plt.show()
```

### Centroid Initialization Methods
We need to find a better initialization strategy.

If you happen to know approximately where the centroids should be (e.g., if you ran another clustering algorithm earlier), 
then you can set the init hyperparameter to a NumPy array containing the list of centroids, and set n_init to 1:

```python
good_init = np.array([[-3, 3], [-3, 2], [-3, 1], [-1, 2], [0, 2]])
kmeans = KMeans(n_clusters=5, init=good_init, n_init=1)
kmeans.fit(X)
plot_decision_boundaries(kmeans, X)
```

Another solution is to run the algorithm multiple times with different random initializations and keep the best solution. 

This is controlled by the ``n_init`` hyperparameter: by default, it is equal to 10, which means that the whole algorithm described earlier actually runs 10 times when you call ``fit()``, and Scikit-Learn keeps the best solution.

**Question**
How can we measure the quality of a solution?

We need a performance metric. 
A very common measure is the  **inertia**.

It is defined as the within-cluster sum-of-squares.


A model’s inertia is accessible via the ``inertia_`` instance variable:

```python
kmeans.inertia_
```

The `score()` method returns the negative inertia. Why negative? Well, it is because a predictor's `score()` method must always respect the "_greater is better_" rule.

```python
kmeans.score(X)
```

K-Means dipende molto da dove metto i centroidi all'inizio.

Se li metto bene, ottengo una buona soluzione.

Se li metto male, potrei ottenere una soluzione scarsa.

Dobbiamo trovare una strategia migliore per inizializzare i centroidi,pk l’inizializzazione casuale può causare risultati diversi e a volte non ottimi globalmente.

Una **soluzione** può essere impostare i centroidi a mano,se si sa già più o meno dove dovrebbero stare i centroidi, posso passarli direttamente a K-Means, quindi invece di fare: init="random" , posso impostare: init=good_init, dove good_init è un array NumPy contenente le coordinate iniziali dei centroidi.

Un’altra **soluzione** è eseguire l’algoritmo più volte con inizializzazioni casuali diverse e tenere la soluzione migliore.

Invece di fidarsi di una sola partenza casuale, faccio:

partenza casuale 1 → soluzione 1
partenza casuale 2 → soluzione 2
partenza casuale 3 → soluzione 3
...

Poi scelgo la migliore.

Questo comportamento è controllato dall’iperparametro:n_init che di default vale 10.

**Come misuriamo la qualità di una soluzione?**

Se faccio K-Means 10 volte, come decido qual è la soluzione migliore?

Serve una misura numerica

Una misura molto comune è l’inerzia(in inglese inertia),è definita come:

somma delle distanze quadratiche dei punti dal proprio centroide.

Per ogni punto calcolo la distanza dal centroide del cluster a cui è assegnato, elevo questa distanza al quadrato e sommo per tutti i punti

Quindi:

inertia bassa = punti vicini ai propri centroidi = clustering più compatto

inertia alta = punti lontani dai propri centroidi = clustering meno buono

**Esempio**

Supponiamo un cluster con centroide:C = [0, 0]

e tre punti:

A = [1, 0]

B = [0, 2]

D = [3, 0]

Distanze dal centroide:

A distanza 1

B distanza 2

D distanza 3

Distanze al quadrato:

1² = 1

2² = 4

3² = 9

Inertia del cluster:1 + 4 + 9 = 14

K-Means cerca di minimizzare questa quantità(l'inerzia).

La funzione .score() restituisce l’inertia negativa.

kmeans.score(X) non dà direttamente l’inertia positiva, ma -inertia,

è negativa pk in Scikit-Learn il metodo score() deve seguire la regola: più alto è meglio,ma per l’inertia vale il contrario: più bassa è meglio.

Allora Scikit-Learn restituisce -inertia così una soluzione migliore avrà score più alto.

Esempio:

Soluzione A:inertia = 100,score = -100

Soluzione B:inertia = 50,score = -50

Quale score è più alto? -50 > -100 ,quindi B è migliore.

### Multiple Initializations

Qui si parla di inizializzazioni multiple.

Cioè la strategia:faccio partire K-Means più volte,poi scelgo la soluzione con inertia minore

After we initialize several times the algorithm, we can select the solution with minimum inertia.

For example, here are the inertia of the two "bad" models shown in the previous figure:

```python
kmeans_rnd_init1.inertia_
```

```python
kmeans_rnd_init2.inertia_
```

As you can see, they have a higher inertia than the first "good" model we trained, which means they are probably worse.

### K-Means++

Instead of initializing the centroids completely at random,
it is preferable to initialize 
them using the following algorithm, proposed in a [2006 paper](https://goo.gl/eNUPw6) by David Arthur and Sergei Vassilvitskii.


* Take one centroid $c_1$, chosen uniformly at random from the dataset.
* Take a new center $c_i$, choosing an instance $\mathbf{x}_i$ with probability: 

$$\frac{D(\mathbf{x}_i)^2 } {\sum\limits_{j=1}^{m}{D(\mathbf{x}_j)}^2}$$

where $D(\mathbf{x}_i)$ is the distance between the instance $\mathbf{x}_i$ and the closest centroid that was already chosen. 

This probability distribution ensures that instances that are further away from already chosen centroids are much more likely be selected as centroids.

* Repeat the previous step until all $k$ centroids have been chosen.

The rest of the K-Means++ algorithm is just regular K-Means. 
With this initialization, the K-Means algorithm is much less likely to converge to a suboptimal solution, so it is possible to reduce `n_init` considerably. 

Most of the time, this largely compensates for the additional complexity of the initialization process.

To set the initialization to K-Means++, simply set `init="k-means++"` (this is actually the default):

Viene introdotta una tecnica migliore per inizializzare i centroidi.

Si chiama **K-Means++**

Invece di inizializzare i centroidi completamente a caso, è meglio usare un algoritmo di inizializzazione più intelligente.

Questo algoritmo si chiama K-Means++ e l’obiettivo è evitare di scegliere centroidi iniziali tutti vicini tra loro.

**Primo passo:**

scegli il primo centroide casualmente dal dataset in maniera uniforme, il che significa che ogni punto ha la stessa probabilità di essere scelto,quindi K-Means++ parte comunque con un po’ di casualità.

**Secondo passo:**

scegli un nuovo centroide, ma non completamente a caso.

La probabilità di scegliere un punto dipende dalla sua distanza dai centroidi già scelti.

Formula della probabilità: $$\frac{D(\mathbf{x}_i)^2 } {\sum\limits_{j=1}^{m}{D(\mathbf{x}_j)}^2}$$

D(x_i) è la distanza tra il punto x_i e il centroide già scelto più vicino.

Quindi per ogni punto guardo, quanto è lontano dal centroide già scelto più vicino?

La formula fa sì che i punti più lontani dai centroidi già scelti abbiano più probabilità di essere scelti come nuovi centroidi.

Inizializzando il tutto in questa maniera, il K-Means++ tende a distribuire i centroidi iniziali meglio, il processo si ripete finchè non sono stati inizializzati tutti i k centroidi. Successivamente alla inizializzazione, il K-Means++ si comporta come il K-Means classico.

### Accelerated K-Means

The K-Means algorithm can be significantly accelerated by avoiding many unnecessary distance calculations: this is achieved by exploiting the triangle inequality
(given three points A, B and C, the distance AC is always such that AC ≤ AB + BC) 
and by keeping track of lower and upper bounds for distances between instances and centroids (see this [2003 paper](https://www.aaai.org/Papers/ICML/2003/ICML03-022.pdf) by Charles Elkan for more details).

To use Elkan's variant of K-Means, just set `algorithm="elkan"`. 

Note that it does not support sparse data, so by default, Scikit-Learn uses `"elkan"` for dense data, and `"lloyd"` (the regular K-Means algorithm) for sparse data.

Il **K-Means classico** funziona così:

1. scelgo k centroidi iniziali

2. assegno ogni punto al centroide più vicino

3. ricalcolo ogni centroide come media dei punti assegnati

4. ripeto finché i centroidi non cambiano più

Il passaggio più costoso è assegnare ogni punto al centroide più vicino,perché per ogni punto devo calcolare la distanza da tutti i centroidi.

Se ho m punti e k centroidi,ad ogni iterazione faccio circa m × k calcoli di distanza

Esempio:

2000 punti e 5 centroidi → 10.000 distanze per iterazione
100.000 punti e 50 centroidi → 5.000.000 distanze per iterazione

Quindi su dataset grandi K-Means può diventare pesante.

**Accelerated K-Means** nasce per velocizzare e per evitare di calcolare distanze inutili

Accelerated K-Means non cambia l’obiettivo di K-Means,l’obiettivo resta sempre minimizzare l’inertia,cioè rendere i punti il più vicini possibile al proprio centroide.

Non cambia neanche l’idea generale, ovvero assegnazione dei punti,aggiornare i centroidi,ripetere fino alla convergenza

Cambia solo come vengono ottimizzati i calcoli

K-Means classico = fa tutti i confronti punto-centroide

Accelerated K-Means = evita alcuni confronti quando capisce che sono inutili.

Esempio di Calcoli inutili :

Abbiamo un punto x che è molto vicino al centroide C1, poi ho un altro centroide C2 molto lontano da C1. Se x è vicino a C1 e C2 è molto lontano da C1, è molto probabile che C2 non possa essere più vicino a x rispetto a C1.

Quindi invece di calcolare ogni volta:

distanza(x, C1)

distanza(x, C2)

distanza(x, C3)

distanza(x, C4)

distanza(x, C5)

l’algoritmo accelerato del K-Means prova a dire che alcuni centroidi sono sicuramente troppo lontani,quindi non spreco tempo a calcolare la distanza esatta.

Per fare ciò, si usa la **disuguaglianza triangolare**.

In geometria, se ho tre punti A, B, C, vale: d(A, C) ≤ d(A, B) + d(B, C)

Cioè la distanza diretta tra A e C non può essere maggiore della distanza passando per B.

Ma per Accelerated K-Means è utile anche questa idea: d(A, C) ≥ |d(A, B) - d(B, C)|

Praticamente posso stimare una distanza minima possibile senza calcolarla direttamente.

Applicata a K-Means, se ho che x = punto,C1 = centroide attuale del punto,C2 = altro centroide

Se so che x è vicino a C1 e C1 è molto lontano da C2, allora posso concludere che C2 probabilmente non può essere il centroide più vicino a x e quindi posso evitare il calcolo preciso di: distanza(x, C2).

L'algoritmo usato si chiama Elkan K-Means, sfrutta lower bound ed upper bound per evitare di fare calcoli inutili, come abbiamo visto precedentemente.

Accelerated K-Means non è sempre automaticamente migliore,per funzionare deve mantenere in memoria varie informazioni aggiuntive, come i bound sulle distanze.Quindi può usare più memoria.Inoltre, se i cluster sono molto sovrapposti o i centroidi sono molto vicini tra loro, potrebbe non riuscire a saltare tanti calcoli, quindi il vantaggio può diminuire.

**ESEMPIO**

Facciamo un esempio numerico semplice con 1 punto e 3 centroidi.

Immagina che un punto x sia attualmente assegnato al centroide C1.

x  = punto

C1 = centroide attuale

C2 = altro centroide

C3 = altro centroide

Supponiamo che sappiamo già che distanza(x, C1) = 0.5

Questa distanza è il nostro upper bound.

Perché upper bound?
Perché sappiamo che il punto x ha già un centroide a distanza 0.5, quindi il centroide migliore non può essere peggiore di 0.5.

In pratica upper bound di x = 0.5 vuol dire che x ha già almeno un centroide raggiungibile entro distanza 0.5

Adesso dobbiamo capire se vale la pena confrontare x con C2 e C3.

**Caso 1**: centroide sicuramente troppo lontano

Supponiamo di sapere, usando la disuguaglianza triangolare, che la distanza minima possibile tra x e C2 ≥ 2.0

Questo è un lower bound,vuol dire che il lower bound per C2 = 2.0.Anche nel migliore dei casi, C2 non può essere più vicino di 2.0.

Ora confrontiamo:

upper bound attuale = 0.5

lower bound da C2 = 2.0

Siccome 0.5 < 2.0 ,allora C2 non può essere il centroide più vicino.

Quindi non serve calcolare davvero distanza(x, C2) perché sappiamo già che C2 è peggiore di C1.

Quindi Accelerated K-Means ci dice di saltare il calcolo della distanza da C2

**Caso 2**: centroide forse competitivo

Ora considera C3,supponiamo che il lower bound sia la distanza minima possibile tra x e C3 ≥ 0.3

Quind il lower bound per C3 = 0.3

Confrontiamo:

upper bound attuale = 0.5
lower bound da C3 = 0.3

Qui abbiamo che 0.3 < 0.5

Questo significa che C3 potrebbe essere più vicino di C1, ma non siamo sicuri, però non possiamo escluderlo.

Quindi in questo caso dobbiamo calcolare davvero distanza(x, C3)

Supponiamo che distanza(x, C3) = 0.8. Allora C3 non è migliore, perché 0.8 > 0.5, quindi x resta assegnato a C1.

**Caso 3**: il punto cambia cluster

Sempre con:

upper bound attuale = distanza(x, C1) = 0.5
lower bound da C3 = 0.3

Dobbiamo calcolare davvero distanza(x, C3).

Se scopriamo che distanza(x, C3) = 0.2, allora C3 è migliore di C1, perché 0.2 < 0.5. quindi il punto cambia cluster:

prima x apparteneva al cluster di C1

dopo x appartiene al cluster di C3

E aggiorniamo anche l’upper bound,nuovo upper bound = 0.2,perché ora il centroide più vicino conosciuto è C3.

Esempio completo tabellare

Punto x attualmente assegnato a C1 , d(x, C1) = 0.5, quindi upper bound = 0.5

Ora confronto gli altri centroidi:

Centroide    Lower bound    Confronto con upper bound    Cosa faccio?

C2           2.0            2.0 > 0.5                   salto il calcolo

C3           0.3            0.3 < 0.5                   calcolo la distanza vera

Per C2 non calcolo nulla, perché è sicuramente troppo lontano.

Per C3 devo calcolare, perché potrebbe essere più vicino.

Dove entra la disuguaglianza triangolare?

Facciamo un esempio geometrico.

Supponiamo:

x è assegnato a C1
d(x, C1) = 0.5
d(C1, C2) = 3.0

Vogliamo stimare quanto potrebbe essere distante x da C2.

Usiamo questa idea: d(x, C2) ≥ d(C1, C2) - d(x, C1)

Quindi:

d(x, C2) ≥ 3.0 - 0.5

d(x, C2) ≥ 2.5

Questo significa che lower bound per C2 = 2.5,ma il punto ha già un centroide a distanza di upper bound = 0.5

Data che 2.5 > 0.5, allora C2 non può essere più vicino di C1,quindi salto il calcolo esatto di d(x, C2).

Riassunto secco

L’**upper bound** dice quanto vicino so già che il punto è al suo centroide attuale

Il **lower bound** dice quanto lontano almeno potrebbe essere un altro centroide

**Regola**:

se lower bound > upper bound,allora quel centroide non può vincere,
quindi non calcolo la distanza esatta.

In forma semplice:

upper bound = distanza dal centroide attuale

lower bound = distanza minima possibile da un altro centroide

se altro centroide è sicuramente più lontano,allora lo ignoro

Questa è l’idea chiave di Accelerated K-Means.

Another, more reasonable, approach is to look at the __silhouette score__, which is the mean __silhouette coefficient__ over all the samples in the dataset. 

An instance's silhouette coefficient is equal to 

$$silhouette(x) = \frac{(b(x) - a(x))}{\max(a(x), b(x))}$$

where $a$ is the mean distance to the other instances in the same cluster (it is the _mean intra-cluster distance_). 

More formally, for each cluster $i \in C$, let $C_i$ denote the set of points
assigned to cluster $i$, then for each point $x \in C_i$ we have:

$$
a(x) = \frac{1}{|C_i|} \sum_{x' \in C_i} d(x, x')   
$$


$b(x)$ is the _mean nearest-cluster distance_, that is the mean distance to the instances of the next closest cluster (defined as the one that minimizes $b(x)$, excluding the x's own cluster). 

$$b(x) = \min_{k \neq i} \frac{1}{|C_k|} \sum_{x' \in C_k} d(x,x')$$

The silhouette coefficient can vary between -1 and +1: a coefficient close to +1 means that the instance is well inside its own cluster and far from other clusters, while a coefficient close to 0 means that it is close to a cluster boundary, and finally a coefficient close to -1 means that the instance may have been assigned to the wrong cluster.

Oltre all' Elbow Method, abbiamo uno strumento più informativo che è la **Silhouette Score**, serve a misurare quanto siano buoni i cluster.

La Silhouette Score Guarda due cose:

1-quanto un punto è vicino agli altri punti del suo stesso cluster

2-quanto è lontano dai punti del cluster più vicino diverso dal suo

Quindi misura sia compattezza interna del cluster e sia separazione dagli altri cluster.

Forumla è : $$silhouette(x) = \frac{(b(x) - a(x))}{\max(a(x), b(x))}$$

a(x) = distanza media di x dagli altri punti del suo stesso cluster,se a(x) è piccolo, significa che il punto è ben vicino ai punti del suo cluster.

b(x) = distanza media di x dai punti del cluster più vicino diverso dal suo,se b(x) è grande, vuol dire che il punto è lontano dagli altri cluster.

In generale se il silhouette(x) va vicino a +1 allora il punto x è ben assegnato, se va vicino a zero allora il punto x sta vicino al confine tra due cluster, se va vicino -1 allora il punto x è stato probabilmente assegnato al cluster sbagliato.

L’**inertia** guarda solo quanto i punti sono vicini al proprio centroide

Il **silhouette score** guarda anche quanto i cluster sono separati tra loro

x non è un nuovo punto che devo inserire in un cluster,è un punto che è già stato assegnato ad un cluster dal K-Means, e dopo l’assegnamento io calcolo il suo valore di silhouette per capire se quell’assegnamento ha senso.

An even more informative visualization is given when you plot every instance's silhouette coefficient, sorted by the cluster they are assigned to and by the value of the coefficient. This is called a _silhouette diagram_:

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter

plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)
    
    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i/ k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)

save_fig("silhouette_analysis_plot")
plt.show() 
```

- The vertical dashed lines represent the silhouette score for each choice of k -- the numbero of clusters. 

- When most of the instances in a cluster have a lower coefficient than this score (i.e., if many of the instances stop short of the dashed line, ending to the left of it),  then the cluster is rather bad since this means its instances are much too close to other clusters. 

__Observations__ 
- when k=3 and when k=6, we get bad clusters. 
- when k=4 or k=5, the clusters look pretty good – most instances extend beyond the dashed line, to the right   and closer to 1.0. 
- when k=4, the cluster at index 1 (the third from the top), is rather big,
- when k=5, all clusters have similar sizes, so even though the overall silhouette score from k=4 is slightly     greater than for k=5, it seems like a good idea to use k=5 to get clusters of similar sizes.

### Limits of K-Means

Despite its many merits, most notably being fast and scalable, K-Means is not perfect.

As we saw, it is necessary to run the algorithm several times to avoid sub-optimal solutions, plus you need to specify the number of clusters, which can be quite a hassle. Moreover, K-Means does not behave very well when the clusters have varying sizes, different densities, or non-spherical shapes. 

For example, the following figure shows how K-Means clusters a dataset containing three ellipsoidal clusters of different dimensions, densities and orientations:

```python
X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
X2 = X2 + [6, -8]
X = np.r_[X1, X2]
y = np.r_[y1, y2]
```

```python
plot_clusters(X)
```

```python
kmeans_good = KMeans(n_clusters=3, init=np.array([[-1.5, 2.5], [0.5, 0], [4, 0]]), n_init=1, random_state=42)
kmeans_pp = KMeans(n_clusters=3, init='k-means++', n_init=10, random_state=42)
kmeans_bad = KMeans(n_clusters=3, random_state=42)

kmeans_good.fit(X)
kmeans_pp.fit(X)
kmeans_bad.fit(X)
```

```python
plt.figure(figsize=(10, 3.2))

plt.subplot(131)
plot_decision_boundaries(kmeans_good, X)
plt.title("Inertia = {:.1f}".format(kmeans_good.inertia_), fontsize=14)

plt.subplot(132)
plot_decision_boundaries(kmeans_bad, X, show_ylabels=False)
plt.title("Inertia = {:.1f}".format(kmeans_bad.inertia_), fontsize=14)

plt.subplot(133)
plot_decision_boundaries(kmeans_pp, X, show_ylabels=False)
plt.title("Inertia = {:.1f}".format(kmeans_pp.inertia_), fontsize=14)

save_fig("bad_kmeans_plot")
plt.show()
```

As you can see, neither of these solutions are any good. 
The solution on the left is better, but it still chops off 25% of the middle cluster and assigns it to the cluster on the right.
The solution on the right is just terrible, even though its inertia is lower. So depending on the data, different clustering algorithms may perform better.

### Using clustering for image segmentation

Image segmentation is the task of partitioning an image into multiple segments. 

In semantic segmentation, all pixels that are part of the same object type get assigned to the same segment. 

For example, in a self-driving car’s vision system, all pixels that are part of a pedestrian’s image might be assigned to the “pedestrian” segment (there would just be one segment containing all the pedestrians). In instance segmentation, all pixels that are part of the same individual object are assigned to the same segment. 

In this case there would be a different segment for each pedestrian. The state of the art in semantic or instance segmentation today is achieved using complex architectures based on convolutional neural networks.

Here, we are going to do something much simpler: color segmentation. We will simply assign pixels to the same segment if they have a similar color. In some applications, this may be sufficient, for example if you want to analyze satellite images to measure how much total forest area there is in a region, color segmentation may be just fine.

For example, it may identify a color cluster for all shades of green. Next, for each color (e.g., dark green), it looks for the mean color of the pixel’s color cluster. 

For example, all shades of green may be replaced with the same light green color (assuming the mean color of the green cluster is light green). Finally it reshapes this long list of colors to get the same shape as the original image. And we’re done!

Now instead of fitting the logistic regressor over the first random 50 points we can use these instances.

Now we have a dataset with just 50 labeled instances, but instead of being completely random instances, each of them is a representative image of its cluster. Let's see if the performance is any better:

```python
log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", random_state=42)
log_reg.fit(X_representative_digits, y_representative_digits)
log_reg.score(X_test, y_test)#dice la accuracy della Logistic Regression sul test set.
#ovvero, quante cifre del test set sono state classificate 
# correttamente dalla Logistic Regression.
```

Wow! We jumped from ~83% accuracy to ~92%, although we are still only training the model on 50 instances. Since it's often costly and painful to label instances, especially when it has to be done manually by experts, it's a good idea to make them label representative instances rather than just random instances.

But perhaps we can go one step further: what if we propagate the labels to all the other instances in the same cluster?

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_==i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", random_state=42)
log_reg.fit(X_train, y_train_propagated)
```

```python
log_reg.score(X_test, y_test)
```

We got a tiny little accuracy boost.

Better than nothing, but we should probably have propagated the labels only to the instances closest to the centroid, because by propagating to the full cluster, we have certainly included some outliers. 

Let's only propagate the labels to the 20th percentile closest to the centroid:

```python
percentile_closest = 20

X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1
```

```python
partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", random_state=42)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
```

```python
log_reg.score(X_test, y_test)
```

Nice! With just 50 labeled instances (just 5 examples per class on average!), we got ~94% accuracy, which is pretty close to the performance of logistic regression on the fully labeled _digits_ dataset (which was 96.7%).

This is because the propagated labels are actually pretty good: their accuracy is very close to 99%:

```python
np.mean(y_train_partially_propagated == y_train[partially_propagated])
```

You could now do a few iterations of _active learning_:
1. Manually label the instances that the classifier is least sure about, if possible by picking them in distinct clusters.
2. Train a new model with these additional labels.

## DBSCAN

This algorithm defines clusters as continuous regions of high density. 

It is actually quite simple:

1. For each instance, the algorithm counts how many instances are located within a small distance $\epsilon$ (epsilon) from it. 
   This region is called the instance’s $\epsilon$-neighborhood.
2. If an instance has at least ``min_samples`` instances in its $\epsilon$-neighborhood (including itself), then it is considered a __core instance__. 
   In other words, core instances are those    that are located in dense regions.
3. All instances in the neighborhood of a core instance belong to the same cluster. 
    This may include other core instances, therefore a __long sequence of neighboring core 
   instances forms a single cluster__.
   
4. Any instance that is not a core instance and does not have one in its neighborhood is considered an anomaly.

![img/dbscan.png](img/dbscan.png)

This algorithm works well if all the clusters are dense enough, and they are well separated by low-density regions.

The DBSCAN class in Scikit-Learn is as simple to use as you might expect. Let’s test it on the moons dataset.

A differenza del K-Means, il DBSCAN non richiede di specificare prima il numero di cluster,DBSCAN definisce i cluster come regioni continue ad alta densità, trova i cluster guardando dove ci sono tanti punti vicini tra loro.
K-Means tende a fare cluster “attorno a centroidi”,quindi con forme curve può sbagliare.
DBSCAN invece non cerca centroidi ma,segue la densità e per questo può trovare cluster curvi,allungati,non circolari,come le due mezze lune.



```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X, y = make_moons(n_samples=1000, noise=0.05, random_state=42)
plt.scatter(X[:,0], X[:,1])
#X contiene i punti, ogni punto ha due features, quindi ogni riga di X
#è un punto, ed X è una matrice 1000X2

#il K-Means funziona bene con cluster sferici,circolari,
# compatti attorno ad un centroide, ma questi punti a forma di luna soffre,
#invece DBSCAN può funzionare bene, perchè guarda la densità locale dei punti
```

```python
print(X.shape, y.shape)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)

#DBSCAN usa principalmente due parametri:eps e min_sample
#eps indica il raggio del vicinato,per ogni punto, DBSCAN guarda 
# quanti punti ci sono entro distanza eps?
#Quella zona si chiama epsilon-neighborhood,cioè vicinato di raggio epsilon.

#min_samples indica quanti punti devono esserci nel vicinato 
# perché un punto sia considerato core point.
#Un core point è un punto in una zona densa.

#Esempio:

#DBSCAN(eps=0.05, min_samples=5) significa che un punto è core 
# se ha almeno 5 punti entro distanza 0.05, contando anche sé stesso.

#TIPI DI PUNTI
#DBSCAN distingue tre tipi di punti.

#1. Core point

#Un punto è core se ha abbastanza vicini nel raggio eps,cioè se ha un 
# numero di punti nel suo eps-neighborhood >= min_samples
#Questi punti stanno in zone dense.

#2. Border point

#Un punto è border se non ha abbastanza vicini per essere core, 
# però si trova nel vicinato di un core point.
#Quindi appartiene comunque a un cluster.

#3. Noise point / anomaly

#Un punto è rumore/anomalia se non è core e 
# non è nel vicinato di nessun core point
#DBSCAN lo etichetta con -1

```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

```python
plt.figure(figsize=(9, 3.2))

plt.subplot(121)
plot_dbscan(dbscan, X, size=100)

plt.subplot(122)
plot_dbscan(dbscan2, X, size=600, show_ylabels=False)

save_fig("dbscan_plot")
plt.show()
```

### Agglomerative Clustering

Agglomerative clustering: a hierarchy of clusters is built from the bottom up. 

Think of many tiny bubbles floating on water and gradually attaching to each other until there’s just one big group of bubbles. 

Similarly, at each iteration agglomerative clustering connects the nearest pair of clusters (starting with individual instances). 

If you draw a tree with a branch for every pair of clusters that merged, you get a binary tree of clusters, where the leaves are the individual instances. 

This approach scales very well to large numbers of instances or clusters, it can capture clusters of various shapes, it produces a flexible and informative cluster tree instead of forcing you to choose a particular cluster scale, and it can be used with any pairwise distance. 

It can scale nicely to large numbers of instances if you provide a connectivity matrix. 

This is a sparse m by m matrix that indicates which pairs of instances are neighbors (e.g., returned by ``sklearn.neighbors.kneighbors_graph()``). Without a connectivity matrix, the algorithm does not scale well to large datasets.

```python
from sklearn.cluster import AgglomerativeClustering
```

Honorable Mention
---
* Birch: this algorithm was designed specifically for very large datasets, and it can be faster than batch K-Means, with similar results, as long as the number of features is not too large (<20). It builds a tree structure during training containing just enough information to quickly assign each new instance to a cluster, without having to store all the instances in the tree: this allows it to use limited memory, while handle huge datasets.

* Affinity propagation: this algorithm uses a voting system, where instances vote for similar instances to be their representatives, and once the algorithm converges, each representative and its voters form a cluster. This algorithm can detect any number of clusters of different sizes. Unfortunately, this algorithm has a computational complexity of O(m2), so it is not suited for large datasets

* Mean-shift: this algorithm starts by placing a circle centered on each instance, then for each circle it computes the mean of all the instances located within it, and it shifts the circle so that it is centered on the mean. Next, it iterates this mean-shift step until all the circles stop moving (i.e., until each of them is centered on the mean of the instances it contains). This algorithm shifts the circles in the direction of higher density, until each of them has found a local density maximum. Finally, all the instances whose circles have settled in the same place (or close enough) are assigned to the same cluster. This has some of the same features as DBSCAN, in particular it can find any number of clusters of any shape, it has just one hyperparameter (the radius of the circles, called the bandwidth) and it relies on local density estimation. However, it tends to chop clusters into pieces when they have internal density variations.
