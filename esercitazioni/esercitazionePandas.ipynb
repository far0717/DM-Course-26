DATA FRAME
è un tabella bidimensionale

```python
import pandas as pd

df = pd.DataFrame({
    "nome": ["Luca","Raffaele","Gianfranmarcangelopierantongiulio"],
    "età": [12,22,647],
    "città": ["Palermo","Cosenza","Roma"],
    "squadra":["Palermo","Inter","Roma"]
})

print(df)
hrule = lambda x=50 : "\n"+ "="*x + "\n"
```

Un Data Frame è composto da Series allineate, ogni colonna del Data Frame è una Series, una Series è un array monodimensionale con indice(labels)

```python
#posso creare un Data Frame anche passando un dato in formato JSON
import pandas as pd
dati={
    "nome": ["Luca","Raffaele","Gianfranmarcangelopierantongiulio"],
    "età": [12,22,647],
    "città": ["Palermo","Cosenza","Roma"],
    "squadra":["Palermo","Inter","Roma"]
}
df = pd.DataFrame(dati)
print(df)
print(hrule())
#se volessi modificare l'indice del mio DataFrame df, basta fare df.index = [nuovo indice]
df.index = ["primo","secondo","terzo"]
print(df)
print(hrule())

#se voglio prendere una o più righe del mio Data Frame, posso uasre i metodi .loc e .iloc, .loc funziona
#con etichette, mentre .iloc con posizioni numeriche
#df.iloc[0] e df.loc["primo"] producono lo stesso risultato
print(df.iloc[0])
print(hrule())
print(df.loc["primo"])
print(hrule())
#se volessi prendere più righe posso usare lo slicing, quindi fare 
print("prendo prima e seconda riga")
print(df.loc["primo":"seconda"]) #con .loc la fine è inclusa
print(hrule())
print("prendo prima e seconda riga")
print(df.iloc[0:2]) # con .iloc la fine è esclusa
print(hrule())
#la struttura generale per .loc e .iloc è 
#df.iloc[righe, colonne]
#df.loc[righe, colonne]

#se voglio prendere una o più colonne(una colonna è una Series), posso fare:
print(df["nome"]) #così con la singola parentesi quadra sto prendendo una Series
print(hrule())
#se voglio prende più colonne,ovvero prendere più Series, ovvero prendere un Data Frame, faccio:
print(df[["nome","squadra"]])
print(hrule())
#se voglio prendere più colonne facendo slicing devo usare .loc oppure .iloc
print(df.loc[:,"nome":"squadra"])
print(hrule())

#Per aggiungere una tupla posso fare df.loc[len(df)] = ["Xenia",21,"Paola","Paola"],
#non posso usare .iloc pk esso funziona solo su posizioni già esistenti,
#però nel nostro caso è meglio fare
df.loc["quarto"] = ["Xenia",22,"Paola","Paola"] # per rispettare il concetto dell'indice
print(df)
print(hrule())
#generalmente conviene lasciare l'indice numerico di default, 
#altrimenti devo aggiungere ogni volta a mano il nome dell'indice di riga quando la creo, come ho fatto adesso
#un altro modo per aggiungere una riga o fondere due Data Frame è :
nuova_riga = pd.DataFrame([["Marco", 22,"Castrovillari","Castrovillari"]], columns=df.columns)

df = pd.concat([df, nuova_riga], ignore_index=True)
print(df)
```
