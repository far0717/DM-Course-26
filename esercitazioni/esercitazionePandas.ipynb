DATA FRAME
è un tabella bidimensionale

```python
import pandas as pd

df = pd.DataFrame({
    "nome": ["Luca","Raffaele","Gianfranmarcangelopierantongiulio"],
    "età": [12,22,647],
    "città": ["Palermo","Cosenza","Roma"],
    "squadra":["Palermo","Inter","Roma"]
})

print(df)
```

Un Data Frame è composto da Series allineate, ogni colonna del Data Frame è una Series, una Series è un array monodimensionale con indice(labels)
