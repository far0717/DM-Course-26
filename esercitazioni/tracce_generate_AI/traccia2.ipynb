```python
# ESERCIZIO 1

from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import (
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    precision_recall_curve
)
import matplotlib.pyplot as plt

data = load_digits()

X = data.data
y = data.target

y_7 = (y == 7)

X_train, X_test, y_train_7, y_test_7 = train_test_split(
    X,
    y_7,
    test_size=0.2,
    random_state=42
)

sgd_clf = SGDClassifier(max_iter=1000, tol=1e-3, random_state=42)

# Predizioni di classe tramite cross validation
y_train_pred = cross_val_predict(
    sgd_clf,
    X_train,
    y_train_7,
    cv=3
)

cm = confusion_matrix(y_train_7, y_train_pred)

precision = precision_score(y_train_7, y_train_pred)
recall = recall_score(y_train_7, y_train_pred)
f1 = f1_score(y_train_7, y_train_pred)

print("Confusion Matrix:")
print(cm)

print("Precision:", precision)
print("Recall:", recall)
print("F1:", f1)

# Score continui per curve e ROC AUC
y_scores = cross_val_predict(
    sgd_clf,
    X_train,
    y_train_7,
    cv=3,
    method="decision_function"
)

precisions, recalls, thresholds = precision_recall_curve(
    y_train_7,
    y_scores
)

plt.plot(thresholds, precisions[:-1], label="Precision")
plt.plot(thresholds, recalls[:-1], label="Recall")
plt.xlabel("Threshold")
plt.legend()
plt.show()

roc = roc_auc_score(y_train_7, y_scores)

print("ROC AUC:", roc)
```

```python
# ESERCIZIO 2

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.multiclass import OneVsOneClassifier
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix

# Creazione pipeline
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("classificatore", OneVsOneClassifier(SVC(gamma="scale")))
])

# Cross validation accuracy
scores = cross_val_score(
    pipeline,
    X,
    y,
    cv=3,
    scoring="accuracy"
)

print("Scores:", scores)
print("Mean accuracy:", scores.mean())

# Predizioni con cross validation
y_pred = cross_val_predict(
    pipeline,
    X,
    y,
    cv=3
)

# Confusion matrix multiclass
cm = confusion_matrix(y, y_pred)

print("Confusion Matrix:")
print(cm)
```

```python
#Esercizio 3
import torch
import torch.nn as nn
import torch.nn.functional as F

class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork,self).__init__()
        self.fc1 = nn.Linear(64,32)
        self.fc2 = nn.Linear(32,16)
        self.fc3 = nn.Linear(16,10)
    def forward(self,x):
        x = x.view(x.shape[0],-1)
        x= F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=F.softmax(self.fc3(x),dim=1)
        #l’output ha shape:(batch_size, 10),quindi
        #  la Softmax va fatta sulle 10 classi, quindi sulla dimensione 1.
        return x
    
```
