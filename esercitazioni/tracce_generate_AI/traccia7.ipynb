```python
# ESERCIZIO 1

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X, y = make_moons(
    n_samples=1000,
    noise=0.05,
    random_state=42
)

dbscan = DBSCAN(eps=0.05, min_samples=5)

dbscan.fit(X)

labels = dbscan.labels_

print("Labels:")
print(labels)

# Numero di outlier, cioè punti con label -1
n_outliers = np.sum(labels == -1)

print("Number of outliers:", n_outliers)

# Indici dei core samples
core_indices = dbscan.core_sample_indices_

print("Core sample indices:")
print(core_indices)

# Coordinate dei core samples
core_points = dbscan.components_

print("Core points:")
print(core_points)
```

```python
# ESERCIZIO 2

import numpy as np

from sklearn.neighbors import KNeighborsClassifier

# Feature dei core points
X_core = dbscan.components_

# Label dei core points
y_core = dbscan.labels_[dbscan.core_sample_indices_]

knn = KNeighborsClassifier(n_neighbors=50)

knn.fit(X_core, y_core)

X_new = np.array([
    [-0.5, 0],
    [0, 0.5],
    [1, -0.1],
    [2, 1]
])

y_new_pred = knn.predict(X_new)

print("Predicted labels:")
print(y_new_pred)

# Distanza dal vicino più vicino
distances, indices = knn.kneighbors(X_new, n_neighbors=1)

print("Distances:")
print(distances)

# Se la distanza è troppo grande, considero il punto outlier
threshold = 0.2

y_new_pred_with_outliers = y_new_pred.copy()

y_new_pred_with_outliers[distances.ravel() > threshold] = -1

print("Predicted labels with outliers:")
print(y_new_pred_with_outliers)
```

```python
#Esercizio 3
import torch
import torch.nn as nn
import torch.nn.functional as F

class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork,self).__init__()
        self.fc1 = nn.Linear(32,24)
        self.fc2 = nn.Linear(24,12)
        self.fc3 = nn.Linear(12,4)
    def forward(self,x):
        x=x.view(x.shape[0],-1)
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=F.softmax(self.fc3(x))
        return x
```
